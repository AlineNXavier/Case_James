## Visualizações


```python
colunas =  ['Order Date', 'Ship Date', 'Ship Mode',
            'Segment', 'City', 'State', 'Country',
            'Market', 'Region', 'Category',
            'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount',
            'Profit', 'Shipping Cost', 'Order Priority']
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df_view = df[colunas]
```

```python
for i, col in enumerate(df_view.select_dtypes(object).columns):
    plt.figure(i)
    sns.countplot(x=col, data=df_view)
```

```python
col_analise = ['Ship Mode', 'Segment', 'Market', 'Region',
               'Category', 'Sub-Category', 'Order Priority',
               'Discount','Quantity','Profit','Shipping Cost']
```

```python
for i,col in enumerate(col_analise):
    plt.figure(i)
    df2 = pd.get_dummies(df[[col,'Sales']])
    sns.heatmap(df2.corr(),annot=True)
```

```python
col_int = ['Sales', 'Quantity','Discount','Profit','Shipping Cost']
# Analise univariavel
for coluna in col_int:
    print(coluna)
    df[coluna].plot.hist(bins = 10, log= True)
    plt.show()
```
